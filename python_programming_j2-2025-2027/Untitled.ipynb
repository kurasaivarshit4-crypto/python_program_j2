````python
"""

write a program to find the student grade based on marks 
1.if student marks is between 100 to 80 then grade A
2.if student marks is between 79 to 60 then grade B
3.if student marks is between 59 to 40 then grade C
4.if student marks is between 29 to 25 then grade D
5.if student marks is between 24 to 0 then grade fail
6.if other then above marks expected output should be please enter correct marks!
"""
marks=int(input("enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks,"A grade")
elif (marks>=60 and marks<=70):
    print(marks,"B grade")
elif (marks>=40 and marks<=59):
    print(marks,"C grade")
elif (marks>=25 and marks<=39):
    print(marks,"D grade")
elif (marks>=0 and marks<=24):
    print(marks,"is fail")
else:
    print(marks,"is not correct marks,please enter correct marks")
````

````python
# write a program to find the largest number fron three numbers
# Program to find the largest of three numbers

# Input: three numbers from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# Compare the numbers
if num1 >= num2 and num1 >= num3:
    largest = num1
elif num2 >= num1 and num2 >= num3:
    largest = num2
else:
    largest = num3

# Output the result
print("The largest number is:", largest)
largest = max(num1, num2, num3)
print("The largest number is:", largest)
````

````python
#write a progeeam to find the leap year or non leap year
# Program to check if a year is a leap year

def is_leap_year(year):
    if (year % 4 == 0):
        if (year % 100 == 0):
            if (year % 400 == 0):
                return True
            else:
                return False
        else:
            return True
    else:
        return False

# Input from user
year = int(input("Enter a year: "))

# Output result
if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
````

````python
# write a program to check login credential using conditional statements
# Predefined username and password
correct_username = "admin"
correct_password = "1234"

# Ask user for input
username = input("Enter your username: ")
password = input("Enter your password: ")

# Check credentials using conditional statements
if username == correct_username and password == correct_password:
    print("Login successful!")
elif username != correct_username and password != correct_password:
    print("Both username and password are incorrect.")
elif username != correct_username:
    print("Incorrect username.")
else:
    print("Incorrect password.")
````
